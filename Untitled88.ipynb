```python
import os
import joblib
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import LabelEncoder

# -------- CONFIG --------
BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Ad Insight"
TRAIN_CSV = os.path.join(BASE_DIR, "archive", "train.csv")

# -------- PREPROCESS --------
def preprocess(df):
    df = df.copy()
    cat_cols = df.select_dtypes(include=["object"]).columns
    for c in cat_cols:
        df[c] = df[c].fillna("UNK")
        le = LabelEncoder()
        df[c] = le.fit_transform(df[c])
    return df.fillna(0)

# Load train data
train_df = pd.read_csv(TRAIN_CSV)
train_df = preprocess(train_df)

# Auto-detect targets (categorical with few unique values)
possible_targets = [
    col for col in train_df.columns
    if train_df[col].nunique() < 20 and col.lower() not in ["id"]
]
feature_cols = [c for c in train_df.columns if c not in possible_targets]

X = train_df[feature_cols].values

print(f"[INFO] Targets detected: {possible_targets}")

acc_scores = {}

for target in possible_targets:
    y = train_df[target].values
    model_path = os.path.join(BASE_DIR, f"{target.lower()}_hybrid_model.pkl")

    if not os.path.exists(model_path):
        print(f"[WARN] No trained model for {target}, skipping.")
        continue

    # Load model tuple (clf, selected_features, best_params)
    clf, feats, params = joblib.load(model_path)

    # Use selected features
    Xsel = X[:, feats]
    y_pred = clf.predict(Xsel)
    acc = accuracy_score(y, y_pred)
    acc_scores[target] = acc

    # ---- Confusion Matrix ----
    cm = confusion_matrix(y, y_pred)
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
    plt.title(f"{target} Confusion Matrix (Accuracy={acc:.2f})")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.tight_layout()
    plt.show()

# ---- Accuracy Bar Chart ----
if acc_scores:
    plt.figure(figsize=(6, 4))
    sns.barplot(x=list(acc_scores.keys()), y=list(acc_scores.values()), palette="viridis")
    plt.ylim(0, 1)
    plt.ylabel("Accuracy")
    plt.title("Hybrid Model Accuracies")
    for i, v in enumerate(acc_scores.values()):
        plt.text(i, v + 0.02, f"{v:.2f}", ha="center", fontsize=10)
    plt.tight_layout()
    plt.show()
```
